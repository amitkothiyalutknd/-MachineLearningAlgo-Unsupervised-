```python
csvdf['Description'] = csvdf['Description'].str.strip()
csvdf.dropna(axis = 0, subset = ['InvoiceNo'], inplace = True)
csvdf = csvdf[-csvdf['InvoiceNo'].str.contains('c')]
```

```python
# csvdf.Country.unique()
csvdf['Country'].unique()
```

```python
basket_Australia = (csvdf[csvdf['Country'] == 'Australia'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
def hot_encode(x):
    if(x <= 0):
        return 0
    if(x >= 1):
        return 1
```

```python
basket_encoded = basket_Australia.applymap(hot_encode)
basket_Australia = basket_encoded
basket_Australia.head()
```

```python
freq_items = apriori(basket_Australia, min_support = 0.1, use_colnames = True)
rules = association_rules(freq_items, metric = 'lift', min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending = [False, False])
print(rules.head())
```

```python
itemlist1 = []
for i in range(len(mso)):
    itemlist1.append([str(mso.values[i,j]) for j in range(20) if str(mso.values[i,j]) != 'nan'])
itemlist1
```

```python
from apyori import apriori
asoc_rules = apriori(itemlist1, min_length = 2, min_support = 0.003, min_confidence = 0.35, min_lift = 3)
asoc_rules = list(asoc_rules)
asoc_rules
```

```python
results = ps.DataFrame(asoc_rules)
results
```

```python
results.nlargest(n = 20, columns = 'support')
```
